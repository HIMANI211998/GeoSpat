## Dealing with files without data

```python
df=pd.read_csv('Apple_no_dates.csv')
df
```

## So my dataframe is pretty bas without dates. How do I generate dates?


```python
rng=pd.date_range(start='2017-06-01',end='2017-06-30',freq='B')  ## freq B for business days.ie exclude weekends
rng
```

## Now set these dates as index of the data frame 

```python
df.set_index(rng,inplace=True)
df.head()
```

```python
df.Close.plot()
```

```python
df.loc['2017-06-01':'2017-06-10'].Close.mean()
```

## How to change frequency? What is you want prices on weekends? on hourly basis? on weekly basis?

```python
df.asfreq('D',method='pad') #Here you are using padding method that is you are carrying forward the prices on weekends
```

```python
df.asfreq('H',method='pad')  #Hourly
```

## What if I know starting date and not ending date but know no. of Period

```python
rng=pd.date_range(start='2017-01-01',freq='B',periods=22)
rng
```

```python
df.set_index(rng,inplace=True)
df
```

## This is very useful when I want to generate random test data

```python
import numpy as np
a=np.random.randint(1,10,len(rng))
a
```

## How do I generate time series out of above data?

```python
ts=pd.Series(a,index=rng)
ts
```
